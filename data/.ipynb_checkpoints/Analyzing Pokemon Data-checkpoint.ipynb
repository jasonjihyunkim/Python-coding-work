```python
import pandas as pd
```

```python
df = pd.read_csv('pokemon_data.csv')
```

```python
df.head()
```

```python
df.describe()
#The values for 'count' aren't important but the mean, std, etc may be useful when comparing with different groups later.
```

```python
df.columns
```

```python
#1) Make a total column, which represents the sum of the numerical attributes.
```

```python
df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']
```

```python
#Creating a checkpoint
df_1 = df.copy(deep=True)
```

```python
df = df_1
```

```python
#This 'Total' column will be placed to the right of 'Type 2' for easy look-up.
```

```python
cols = list(df.columns.values)
df = df[cols[0:4] + [cols[-1]] + cols[4:12]]
```

```python
df
```

```python
#2) Finding the Type of pokemon with the highest HP (average)
```

```python
HP_table = df.groupby(['Type 1']).mean().sort_values('HP',ascending = False)['HP']
#From this, we can see that Dragon Types have the highest average HP.
```

```python
#3) We can compare how the HP of various Types compare to the overall average HP of all pokemon
```

```python
df['HP'].mean()
```

```python
#4) Count how many pokemon are in each type
df.groupby(['Type 1']).count()['#'].sort_values(ascending = False)
#We can see that there are most number of Water Type pokemons
```

```python
#5) We can find out how many 'Mega' pokemon there are 
mega_pokemon = df.loc[df['Name'].str.contains('Mega')]
count = mega_pokemon.count()['#']
print(f"There are {count} mega pokemon that are currently known.")
```

```python
#6) Get a ranking of the best pokemon by sorting them by their total
best_pokemon = df.sort_values('Total', ascending = False)
best_pokemon.reset_index(drop = True, inplace = True)
best_pokemon.index += 1
best_pokemon
```

```python
#7) We're going to export this data into an excel file for future use
best_pokemon.to_csv('pokemon_ranked.csv')
```
