```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline
```

```python
#First, a copy was created from the original file to clean up the data
#The column names were initially in the form 'Unnamed: 1, Unnamed: 2, etc'. Using the rename function, they were changed to 
#sensible heading names
NBA_data_clean = NBA_data.copy(deep = True)
NBA_data_clean.columns = ['Rank','Full Name', 'Team', 'Position','Age','Games Played','MPG','Min%','USG%','TO%','FTA','FT%','2PA','2P%','3PA','3P%','eFG%','TS%','PPG','RPG','TRB%','APG','AST%','SPG','BPG','TOPG','VI','Offensive Rating','Defensive Rating']
```

```python
NBA_data_clean.head()
```

```python
#Excluding the first row which gives detailed description of the headings, this shows that there are 591 players in the NBA this season
#We don't need the Rank column as this information is incomplete and useless to what we wish to accomplish.
#We also exclude the following columns as they're not relevant to what we wish to compare: Min%, USG%, eFG%, TS%, TRB%, AST%, TOPG, VI.
NBA_data_clean.info(verbose = True)
```

```python
NBA_data_clean.info()
```

```python
NBA_data_clean.drop(index = 0, inplace = True)
#This drops the first row which was a duplicate of our column headings.
```

```python
NBA_data_clean.drop(['Rank','USG%','TS%','TRB%','AST%','TOPG','VI'], axis = 1)
```

```python
NBA_data_clean.describe().T
#The data below shows some interesting things such as that there are three 'Jordan McRae' in the NBA currently.
#I want to identify the values that are null (e.g. TO% and eFG% has two missing players)
```

```python
NBA_data_clean.loc[NBA_data_clean['Full Name'] == 'Anthony Tolliver']
#We can see that the 'Anthony Tolliver' are all indeed different people, and are not some duplicate values.
```

```python
NBA_data_clean[pd.isnull(NBA_data_clean["TO%"])]
#The two players missing TO% are also the ones missing eFG%. This is due to lack of games and minutes played.
```

```python
NBA_data_clean[pd.isnull(NBA_data_clean["Offensive Rating"])]
NBA_data_clean[pd.isnull(NBA_data_clean["Defensive Rating"])]
#These players also have no played enough minutes in the season to have an Offensive or Defensive Rating. 
```

```python
#There are many things that we can analyze, but the analysis will have more meaning if we base them on players who qualified
#for enough minutes. These players without enough minutes are outliers that can skew the data. We will omit them from our set.
NBA_data_clean.dropna(how="any", thresh=None, subset=None, inplace=True)
```

```python
NBA_data_clean.info()
#Now we can start our analysis and comparisons of interesting data
NBA_data_clean.to_csv('NBA_data_cleaned.csv')
```

```python
#Effect of age on stats
x = NBA_data_clean[['Age']]
y1 = NBA_data_clean[['PPG']]
y2 = NBA_data_clean[['TO%']]
y3 = NBA_data_clean[['Offensive Rating']]
y4 = NBA_data_clean[['Defensive Rating']]
```

```python
fig, (ax1,ax2) = plt.subplots(2)
fig.set_size_inches(20, 15)
fig.suptitle("Age vs. various statistics", fontsize = 24)
ax1.scatter(x,y1,c='red')
ax1.set_xlabel ("Age")
ax1.set_ylabel ("Points Per Game")
ax2.scatter(x,y2)
ax2.set_xlabel ("Age")
ax2.set_ylabel("Turnover Percentage")
```

```python
fig, (ax3,ax4) = plt.subplots(2)
fig.set_size_inches(20, 15)
fig.suptitle("Age vs. various statistics", fontsize = 24)
ax3.scatter(x,y3,c='green')
ax3.set_xlabel ("Age")
ax3.set_ylabel("Offensive Rating")
ax4.scatter(x,y4,c='black')
ax3.set_xlabel("Age")
ax3.set_ylabel("Defensive Rating")
```

```python
#relating mpg to ppg 
#relating rpg to bpg 
#relating FT% to 2P% and 3P%
#relating TO% to Offensive Rating
```

```python
plt.figure(figsize =(15,15))
x = NBA_data_clean[['MPG']]
y1 = NBA_data_clean[['PPG']]
y2 = NBA_data_clean[['RPG']]
y3 = NBA_data_clean[['APG']]
plt.scatter(x,y1,c='lightblue',label='PPG')
plt.scatter(x,y2,c='pink',label='RPG')
plt.scatter(x,y3,c='purple',label='APG')
plt.legend(loc="upper left", prop={'size':25})
plt.title("Minutes Per Game Vs Various Stats",fontsize = 24)
plt.xlabel("Minutes Per Game", fontsize = 14)
plt.ylabel("Points/Rebounds/Assists Per Game", fontsize = 14)
```

```python
plt.figure(figsize =(15,10))
x = NBA_data_clean[['RPG']]
y = NBA_data_clean[['BPG']]
plt.scatter(x,y,color='blue')
plt.title("Rebounds Per Game VS Blocks Per Game", fontsize = 24)
plt.xlabel("Rebounds Per Game", fontsize = 14)
plt.ylabel("Blocks Per Game", fontsize = 14)
```

```python
plt.figure(figsize =(15,15))
x = NBA_data_clean[['FT%']]
y1 = NBA_data_clean[['2P%']]
y2 = NBA_data_clean[['3P%']]
plt.scatter(x,y1,color='red',label='FT% VS 2P%')
plt.scatter(x,y2,color='green',label='FT% VS 3P%')
plt.legend(loc="upper left", prop={'size':15})
plt.title("Free Throw Percentage VS 2-Point and 3-Point Percentage", fontsize = 24)
plt.xlabel("Free Throw Percentage", fontsize = 14)
plt.ylabel("2-Point/3-Point Percentage", fontsize = 14)

#Modify this so that only data with FTA, 2PA and 3PA higher than certain amounts are included.
```

```python
plt.figure(figsize =(15,10))
x = NBA_data_clean[['TO%']]
y = NBA_data_clean[['Offensive Rating']]
plt.scatter(x,y,color='black')
plt.title("Turnover Percentage VS Offensive Rating", fontsize = 24)
plt.xlabel("Turnover Percentage", fontsize = 14)
plt.ylabel("Offensive Rating", fontsize = 14)
```
